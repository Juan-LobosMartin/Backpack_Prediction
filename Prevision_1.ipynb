```python
# Modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Backpack_df = pd.read_csv('Data/train.csv')
Backpack_df.head(3)
```

```python
Backpack_df.info()
```

```python
Backpack_df.describe()
```

```python
Backpack_df.columns
```

```python
# to see all values in columns
for column in Backpack_df.columns:
    print(f"Value Counts for Column '{column}':")
    print(Backpack_df[column].value_counts(dropna=False))
    print("-" * 30)  # Separator for clarity
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Style', y='Weight Capacity (kg)', data=Backpack_df)
plt.title('Weight max Vs Style')
plt.xlabel('Style')
plt.ylabel('Weight Capacity (kg)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Size', y='Weight Capacity (kg)', data=Backpack_df)
plt.title('Weight max Vs Size')
plt.xlabel('Size')
plt.ylabel('Weight Capacity (kg)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Compartments', y='Price', data=Backpack_df) # jitter adds a bit of random noise to x-axis
plt.title('Price vs. Number of Compartments')
plt.xlabel('Number of Compartments')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Compartments', y='Price', data=Backpack_df)  # Shows mean and confidence intervals
plt.title('Average Price by Number of Compartments')
plt.xlabel('Number of Compartments')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='Compartments', y='Price', data=Backpack_df, marker='o') # Add markers to show data points
plt.title('Price Trend by Number of Compartments')
plt.xlabel('Number of Compartments')
plt.ylabel('Price')
plt.xticks(range(1, 11)) # Make sure all x-axis values are shown.
plt.show()
```
